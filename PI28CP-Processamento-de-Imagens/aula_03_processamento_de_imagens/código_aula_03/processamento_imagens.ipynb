#### Desenhando na imagem

```python
img = cv2.imread('lenna.png')
```

```python
img[50 : 80, :] = (255, 255, 0) #no intervalo da linha 50 até a 80, e em todas as colunas, pinte de azul claro
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img[100 : 120, 100 : 120] = (0, 255, 255) #cria um quadrado amarelo de 100 a 120 na linha e de 100 a 120 na coluna.
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img[170 : 180, 40 : 200] = (255, 255, 255) #cria um retângulo branco de 170 a 180 na linha e de 40 a 200 na coluna.
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### Desenhando e Escrevendo em OpenCV

```python
img = cv2.imread('lenna.jpeg')
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
img.shape
```

#### Desenhando linhas verde e vermelha de espessura 3 na imagem

```python
img1 = img.copy()
```

```python
cv2.line(img1, (100, 0), (100, 224), (0, 255, 0), 3)
cv2.line(img1, (0, 100), (224, 100), (255, 0, 0), 3)
```

```python
plt.imshow(img1)
plt.axis('off')
```

#### Desenhando um círculo na imagem

```python
img1 = img.copy()
```

```python
cv2.circle(img1, (130, 130), 40, (0, 255, 0), 2)
```

```python
plt.imshow(img1)
plt.axis('off')
```

#### Desenhando um retângulo na imagem

```python
img1 = img.copy()
```

```python
cv2.rectangle(img1, (100, 100), (160, 170), (0, 0, 255), 2)
```

```python
plt.imshow(img1)
plt.axis('off')
```

#### Exemplo de escrita de texto

```python
img1 = img.copy()
```

```python
cv2.putText(img1, 'Lenna', (105, 190), cv2.FONT_HERSHEY_PLAIN, 1.5, 
            (255, 255, 0), 2, cv2.LINE_AA)
```

```python
plt.imshow(img1)
plt.axis('off')
```

#### Combinação dos exemplos anteriores

```python
img1 = img.copy()
```

```python
cv2.line(img1, (100, 0), (100, 224), (0, 255, 0), 3)
cv2.line(img1, (0, 100), (224, 100), (255, 0, 0), 3)
cv2.circle(img1, (130, 130), 40, (0, 255, 0), 2)
cv2.rectangle(img1, (100, 100), (160, 170), (0, 0, 255), 2)
cv2.putText(img1, 'Lenna', (105, 190), cv2.FONT_HERSHEY_PLAIN, 1.5, 
            (255, 255, 0), 2, cv2.LINE_AA)
```

```python
plt.imshow(img1)
plt.axis('off')
```
